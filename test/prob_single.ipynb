```python
from qcml import QCML
import numpy as np
import ecos
import scipy.sparse as sp

def cvxopt2numpy(x):
    temp = np.array(x).flatten()
    if len(temp) > 0:
        return temp
    else:
        return None

def cvxopt2sparse(A):
    m,n = A.size
    I = cvxopt2numpy(A.I)
    J = cvxopt2numpy(A.J)
    V = cvxopt2numpy(A.V)
    if m == 0:
        return None
    return sp.coo_matrix((V,(I,J)),shape=(m,n)).tocsc()

def convert_socp(a):
    a['A'] = cvxopt2sparse(a['A'])
    a['G'] = cvxopt2sparse(a['G']) 
    a['c'] = cvxopt2numpy(a['c'])
    a['b'] = cvxopt2numpy(a['b'])
    a['h'] = cvxopt2numpy(a['h'])


class Problem(object):
    def __init__(self):
        m = 10
        n = 20
        A = np.random.randn(m,n)
        b = np.random.randn(m)
        
        q = QCML()
        self.q = q
        q.parse('''
        dimensions m n
        variable x(n)
        parameters A(m,n) b(m)
        minimize norm1(x)
        A*x == b
        ''')
        q.canonicalize()
        q.dims = {'m':m,'n':n}
        q.codegen('python')
        #qcml outputs cvxopt mats
        self.socp_vars = q.prob2socp(locals())
        
        #convert to numpy/scipy.sparse
        convert_socp(self.socp_vars)

        self.ecos_sol = ecos.solve(**self.socp_vars)
        
        #solution to transformed socp
        self.socp_sol = self.ecos_sol['x']
        #solution to original problem
        self.prob_sol = q.socp2prob(self.ecos_sol['x'])['x']
        
prob = Problem()
prob.prob_sol
```
